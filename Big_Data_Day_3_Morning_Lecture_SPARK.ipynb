```
spark = SparkSession.builder.appName(APP_NAME).getOrCreate()

credit = spark.read.format('csv').option('header', 'true').option('inferSchema', 'true').load('/content/drive/MyDrive/DSI Month 4/Week 15/Data/creditcard.csv')
```

```
credit.printSchema()
```

```
credit.show()
```

```
credit.take(2)
```

```
credit_v1 = credit.withColumnRenamed('V1', 'Variable1')

credit_v1.show()
```

```
from pyspark.sql.types import StringType

credit_str = credit_v1.withColumn('Class_Str', credit_v1['Class'].cast(StringType()))

credit_str.printSchema()
```

```
credit_str.show()
```

```
credit_amount = credit_str.withColumn('AmountCents', credit_str['Amount'] * 100)
```

```
credit_amount.select('AmountCents', 'Amount').show()
```

```
from pyspark.sql.functions import UserDefinedFunction, udf
from pyspark.sql.types import StringType, IntegerType, DoubleType, BooleanType

accept_reject_udf = UserDefinedFunction(lambda x: 'Accept' if x==1 else 'Reject', StringType())
# accept_reject_udf = udf(lambda x: 'Accept' if x==1 else 'Reject', StringType())

credit_string = credit.withColumn('ClassStr', accept_reject_udf(credit['Class']))
credit_string.select('Class', 'ClassStr').show()
```

```
credit_string.registerTempTable('credit')
greaterThan1000 = spark.sql("SELECT * FROM credit WHERE Amount > 1000")

greaterThan1000.show()
```

```
greaterThan1000.select('Amount').show()
```

```
df = credit.groupBy('Class').mean('Amount').toPandas()
```
