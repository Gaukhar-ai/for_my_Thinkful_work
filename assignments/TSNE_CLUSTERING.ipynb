```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
#load the dataset and make the preprocessing(normalizing)
```

```python
mnist=fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(123)
#select 10000 images randomly
indices = np.random.choice(3000, 50000)
X = mnist.data[indices]/255.0
y = mnist.target[indices]

print(X.shape, y.shape)
```

```python
#apply tSNE to the data by setting n_components=2
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(X)
```

using the two-dimensional t-SNE representation, draw a graph of the data by coloring and labeling the data points as we did in the checkpoint.

```python
plt.figure(figsize=(10,5))
colours = ['r', 'b', 'g', 'c', 'm', 'y', 'k', 'r', 'burly wood', 'chartreuse']
for i in range(tsne_results.shape[0]):
    plt.text(tsne_results[i, 0], tsne_results[i, 1], str(y[i]),
            color=colours[int(y[i])],
            fontdict={'weight': 'bold', 'size': 50})
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

does the t-sne solution is satisfactory? can i distinguish between different classes easily? which one has done a better job: t_SNE or the PCA i applied above?

```python
plt.figure(figsize=(10, 5))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1])
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

by using t-SNE, we can distinguish some classes clearly and in this respect t-SNE solution is better than that of PCA's 

now, play with the different perplexity values and apply t-SNE for each of them. Which perplexity value is the best one in terms of the two-dimensional representation clarity?

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=10, n_iter=300)
tsne_results = tsne.fit_transform(X)

plt.figure(figsize=(10, 5))
plt.title('t-SNE with perplexity=10')
plt.scatter(tsne_results[:, 0], tsne_results[:, 1])
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=20, n_iter=300)
tsne_results = tsne.fit_transform(X)

plt.figure(figsize=(10, 5))
plt.title('t-SNE with perplexity=20')
plt.scatter(tsne_results[:, 0], tsne_results[:, 1])
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```


All the different configurations of perplexity achieves better results than PCA, however choosing one of them isn't that much easy by just looking at unlabeled and uncolored graphs.
