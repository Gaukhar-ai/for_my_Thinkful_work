```python
# Get Data# Parameters 
n = 5 #the # of article headlines displayed per ticker
tickers = ['SNP', 'PTR', 'BP', 'XOM', 'TOT', 'CVX', 'MPC']

#particularly selected OIL mining companies, since the market is very unique these times,
#lots of opportunities in the oil market
finviz_url = 'https://finviz.com/quote.ashx?t='
news_tables = {}
```

```python
#cleaning the data by using BeautifulSoup
d = {'ticker': [], 'news':[], 'date': []}
df_news = pd.DataFrame(data=d)


for ticker in tickers:
    url = finviz_url + ticker
    req = Request(url=url,headers={'user-agent': 'my-app/0.0.1'}) 
    resp = urlopen(req)    
    html = BeautifulSoup(resp, features="lxml")
    news_table = html.find(id='news-table')
    news_tables[ticker] = news_table
    #print(news_table)

try:
    for ticker in tickers:
        df = news_tables[ticker]
        df_tr = df.findAll('tr')
    
        print ('\n')
        #print ('Recent News Headlines for {}: '.format(ticker))
        
        for i, table_row in enumerate(df_tr):
            a_text = table_row.a.text
            td_text = table_row.td.text
            td_text = td_text.strip()
            
            d = {'ticker': ticker, 'news': a_text, 'date': td_text}
            df_news = df_news.append(d, ignore_index=True)
            #print(a_text,'(',td_text,')')
            #print(td_text)
            #print('---------')
            if i == n-1:
                break
except KeyError:
    pass
```

```python
def remove_punc(test_str):
    # initializing punctuations string  
    punc = '''!()-[]{};:'"\,<>/?@#$%^&*_~'''
      
    # Removing punctuations in string 
    # Using loop + punctuation string 
    for ele in test_str:  
        if ele in punc:  
            test_str = test_str.replace(ele, "") 
    return (test_str)

for i in range(0,len(df_news)):
    #print(i)
    pattern = "[\[].*?[\]]"
    #remove prantheses
    df_news['news'][i] = re.sub(pattern, "", df_news['news'][i])  #remove special characters
    df_news['news'][i] = ' '.join((df_news['news'][i]).split())   #removeextra whitespace, newlines
    df_news['news'][i] = remove_punc(df_news['news'][i])
    df_news['news'][i] = ''.join([i for i in df_news['news'][i] if not i.isdigit()])
    
    #print(df_news['news'][i])
    
#remove duplicate articles
#-------------------------

nlp      = spacy.load('en')
news_doc = nlp(df_news['news'].str.cat(sep=' '))
    
```

```python
# Utility function to calculate how frequently words appear in the text
def word_frequencies(text):
    
    # Build a list of words
    # Strip out punctuation
    words = []
    for token in text:
        if not token.is_punct:
            words.append(token.text)
            
    # Build and return a `Counter` object containing word counts
    return Counter(words)

# Instantiate your list of the most common words
alice_word_freq = word_frequencies(news_without_stopwords).most_common(10)
```

```python
print(alice_word_freq)
```

```python
#Lemmatization Utility function to calculate how frequently each lemma appears in the text
def lemma_frequencies(text):
    
    # Build a list of lemmas
    # Strip out punctuation
    lemmas = []
    for token in text:
        if not token.is_punct:
            lemmas.append(token.lemma_)
            
    # Build and return a `Counter` object containing lemma counts
    return Counter(lemmas)

# Instantiate your list of most common lemmas
news_lemma_freq = lemma_frequencies(news_without_stopwords).most_common(20)
```
